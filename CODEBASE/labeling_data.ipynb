```python
def label_score(score):
    return -1.0 if score <= 2 else \
           0.0 if score == 3 else 1.0

X = [x for x in df_tiktok.content]
Y = [label_score(score) for score in df_tiktok.score]

label = []
for x in df_tiktok.score:
    label = Y

# print(label)
print(f"sample of labeled data")
print(f"{Y[10]}: {X[10][0:100]}...")
print(f"{Y[920]}: {X[920][0:100]} ...")
```

```python
df_tiktok['label'] = label
df_tiktok.head()
```
